```python
import matplotlib.pyplot as plt
from typing import List, Dict, Callable
import random
```

```python
%matplotlib inline
```

```python
from utils import measure_time_complexity, measure_space_complexity
from bubble_sort import bubble_sort
from insertion_sort import insertion_sort
from selection_sort import selection_sort
```

```python
# Define the algorithms to be tested
algorithms: Dict[str, Callable[[List[int]], List[int]]] = {
    "Bubble Sort": bubble_sort,
    "Insertion Sort": insertion_sort,
    "Selection Sort": selection_sort,
}
```

```python
# Define the list sizes to be tested
list_sizes: List[int] = [10, 100, 1000, 10000]
```

```python
time_complexities = {name: [] for name in algorithms.keys()}
space_complexities = {name: [] for name in algorithms.keys()}
```

```python
# For each list size
for size in list_sizes:
    # Generate a random list of that size
    list = random.sample(range(size), size)

    # For each algorithm
    for name, algorithm in algorithms.items():
        # Measure time complexity
        time_complexity = measure_time_complexity(algorithm, list.copy())
        time_complexities[name].append(time_complexity)

        # Measure space complexity
        space_complexity = measure_space_complexity(algorithm, list.copy())
        space_complexities[name].append(space_complexity)
```

```python
# Time Complexity Plot
for name, times in time_complexities.items():
    plt.plot(list_sizes, times, label=name)
plt.xlabel('List Size')
plt.ylabel('Time Complexity')
plt.legend()
plt.title('Time Complexity of Sorting Algorithms')
plt.grid(True)
```

```python
# Space Complexity Plot
for name, spaces in space_complexities.items():
    plt.plot(list_sizes, spaces, label=name)
plt.xlabel('List Size')
plt.ylabel('Memory Complexity')
plt.title('Space Complexity of Sorting Algorithms')
plt.legend()
plt.grid(True)
```
